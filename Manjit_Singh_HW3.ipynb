# Take the pandas dataset and split it into our features (X) and label (y)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)

```python
X = slime[["Temperature °C","Mols KCL"]]
y = slime["Size nm^3"]
X_train, X_test, Y_train, Y_test=train_test_split(X,y,test_size=0.1,train_size=0.9)
```

# Use sklearn to train a model on the training set

# Create a sample datapoint and predict the output of that sample with the trained model

# Report on the score for that model, in your own words (markdown, not code) explain what the score means

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX

```python
X_train
```

```python
Y_train
```

```python

reg = LinearRegression()
reg.fit(X_train,Y_train).predict(X_train)
```

```python
reg.fit(X_train,Y_train).score(X_train,Y_train)
```

The Output score is known as the coeffiecient of determination, which is 1-u/v. This score seems to show how good its best fit line is by by showing improvments to the regression's cost by comparing its cost from its current values, vs its cost from its average. The max score it can have is 1.0 as it shows that the regression's cost is 0.

```python
reg.coef_
```

```python
reg.intercept_
```

# Use the cross_val_score function to repeat your experiment across many shuffles of the data
